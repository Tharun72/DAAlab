```python
import time
import random
import matplotlib.pyplot as plt

def linear_search(arr, key):
    for i in range(len(arr)):
        if arr[i] == key:
            return i
    return -1

input_sizes = [1000, 2000, 4000, 6000]
best_times = []
average_times = []
worst_times = []

for n in input_sizes:
    arr = random.sample(range(1, n * 10), n)

    best_case_key = arr[0]
    average_case_key = arr[n // 2]
    worst_case_key = -1

    start = time.time()
    linear_search(arr, best_case_key)
    best_times.append((time.time() - start) * 1000)

    start = time.time()
    linear_search(arr, average_case_key)
    average_times.append((time.time() - start) * 1000)

    start = time.time()
    linear_search(arr, worst_case_key)
    worst_times.append((time.time() - start) * 1000)

plt.plot(input_sizes, best_times, marker='o', label='Best Case', color='green')
plt.plot(input_sizes, average_times, marker='o', label='Average Case', color='blue')
plt.plot(input_sizes, worst_times, marker='o', label='Worst Case', color='red')

plt.title("Linear Search Time Complexity")
plt.xlabel("Input Size (n)")
plt.ylabel("Execution Time (ms)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

```
