```python
import random
import matplotlib.pyplot as plt

input_sizes = [100, 500, 1000, 5000, 10000]
best_case_comparisons = []
average_case_comparisons = []
worst_case_comparisons = []

for n in input_sizes:
    data = list(range(n))
    
    key_best = data[0]
    key_avg = data[n//2]
    key_worst = -1

    comparisons_best = 0
    for i in range(n):
        comparisons_best += 1
        if data[i] == key_best:
            break

    comparisons_avg = 0
    for i in range(n):
        comparisons_avg += 1
        if data[i] == key_avg:
            break

    comparisons_worst = 0
    for i in range(n):
        comparisons_worst += 1
        if data[i] == key_worst:
            break

    best_case_comparisons.append(comparisons_best)
    average_case_comparisons.append(comparisons_avg)
    worst_case_comparisons.append(comparisons_worst)

plt.plot(input_sizes, best_case_comparisons, marker='o', color='green', label='Best Case')
plt.plot(input_sizes, average_case_comparisons, marker='o', color='blue', label='Average Case')
plt.plot(input_sizes, worst_case_comparisons, marker='o', color='red', label='Worst Case')

plt.title("Linear Search - Number of Comparisons")
plt.xlabel("Input Size (n)")
plt.ylabel("Comparisons")
plt.legend()
plt.grid(True)
plt.show()
```
